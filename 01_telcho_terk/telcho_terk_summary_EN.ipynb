# 📘 Telcho Project – Introduction to Machine Learning

This Jupyter notebook belongs to the **"Telcho"** project, which explores the fundamental concepts of machine learning (ML) from both theoretical and practical perspectives. The project aims to help learners understand the basics of supervised learning and recognize its applications across various sectors.

---

## 📌 1. Key Concepts

Machine learning is a subfield of artificial intelligence that enables systems to **learn from data** without being explicitly programmed. This project elaborates on the following concepts:

- **Machine Learning (ML)**: Learning patterns from data
- **Artificial Intelligence (AI)**: Systems capable of decision making
- **Data Science**: The process of extracting knowledge from data

Bu kavramlar arasında farklar, tablo ve örneklerle açıklanmıştır.

---

## 🌐 2. Real-World ML Applications

### 🏦 Finance
- Credit scoring, fraud detection

### 🏥 Healthcare
- Disease diagnosis, drug discovery

### 🛒 E-Commerce
- Recommendation systems, demand forecasting

### 🚗 Autonomous Systems
- Intelligent transportation, sensor data analysis

Her sektör için günlük hayatta karşılaşılabilecek özgün senaryolar üzerinden açıklamalar yapılmıştır.

---

## 🎯 3. Project Objective and Learning Outcomes

Through this project, the student:
- Gains an understanding of how ML models function.
- Develops the ability to apply theoretical knowledge in practice.
- Learns to distinguish between Data Science, AI, and ML.
- Gains insight into real-world data scenarios.
